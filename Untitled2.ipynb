```python
def convolution(image, filterr, bias):
  channel,m,n=image.shape
  numofFilt,channels,f,s=filterr.shape
  outputDim=m-f+1
#   print(f)
  output=np.zeros((numofFilt,outputDim,outputDim))
 
  for count in range(numofFilt):
    for x in range(outputDim):
      for y in range(outputDim):
        output[count,x,y]=np.sum(filterr[count,:,:]*image[:,x:x+f,y:y+f])
  return output    
    
  
  
```

```python
def maxPool(image,f=2,s=2):
  print("Pool ",image.shape)
  channel,m,n=image.shape
  outputDim=int((m-f)/s)+1
  output=np.zeros((channel,outputDim,outputDim))
  for count in range(channel):
    for x in range(outputDim):
      for y in range(outputDim):
        output[count,x,y]=np.max(image[count,s*x:s*x+f,s*y:s*y+f])
  return output    
```
